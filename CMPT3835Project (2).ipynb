```python
from sklearn.preprocessing import MinMaxScaler

# Choose features and target
features = ['lag_7', 'lag_14', 'lag_30', 'rolling_mean_7', 'rolling_std_7',
            'latitude_client', 'longitude_client']
target = 'quantity'

# Scaling
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(df[features])
y = df[target].values

# Train-test split (e.g., last 90 days for test)
split = int(len(df) * 0.8)
X_train, X_test = X_scaled[:split], X_scaled[split:]
y_train, y_test = y[:split], y[split:]
```

```python
#Optional Hybrid Model (LSTM + XGBoost or FCNN + geospatial)
#Use LSTM for time dependency + XGBoost or Dense layers for tabular data

#Use two-branch architecture for hybrid input: sequence + location
# hybrid deep learning model for predicting food hamper quantity, combining LSTM (for temporal dependencies) with geospatial and engineered features.
#1. Data Preparation
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Select important features
features = [
    'lag_7', 'lag_14', 'lag_30',
    'rolling_mean_7', 'rolling_std_7',
    'latitude', 'longitude',
    'pickup_dayofweek', 'pickup_month', 'pickup_quarter', 'pickup_is_weekend',
    'year', 'month', 'week', 'day', 'day_of_week', 'hour', 'is_weekend'
]
target = 'quantity'

# Drop rows with NaNs after lag/rolling creation
df_model = df.dropna(subset=features + [target]).copy()

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_model[features])

# Prepare LSTM sequences (e.g., using window=7)
def create_sequences(X, y, time_steps=7):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

X_seq, y_seq = create_sequences(X_scaled, df_model[target].values)
```

```python
#2. LSTM + Dense Model Architecture
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

model = Sequential([
    LSTM(64, return_sequences=False, input_shape=(X_seq.shape[1], X_seq.shape[2])),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()
```

```python
#  Time Series Model Comparison for Food Hamper Quantity Forecasting
#  Chronological Split (Last 90 days as test set)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from prophet import Prophet
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings("ignore")

# === 1. Prepare Data ===
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')
df = df.drop_duplicates(subset='timestamp')

# Create time-series DataFrame
time_series = df[['timestamp', 'quantity']].copy()
time_series.set_index('timestamp', inplace=True)

# Chronological Split
train = time_series.iloc[:-90].copy()
test = time_series.iloc[-90:].copy()

# Evaluation Function
def evaluate_forecast(true, predicted, model_name):
    rmse = np.sqrt(mean_squared_error(true, predicted))
    mae = mean_absolute_error(true, predicted)
    print(f"{model_name} -> RMSE: {rmse:.2f}, MAE: {mae:.2f}")
    return rmse, mae

results = {}

# === 2. ARIMA ===
model_arima = ARIMA(train['quantity'], order=(5,1,0))
fit_arima = model_arima.fit()
pred_arima = fit_arima.forecast(steps=90)
results['ARIMA'] = evaluate_forecast(test['quantity'], pred_arima, 'ARIMA')

# === 3. SARIMA ===
model_sarima = SARIMAX(train['quantity'], order=(1,1,1), seasonal_order=(1,1,1,12))
fit_sarima = model_sarima.fit()
pred_sarima = fit_sarima.forecast(steps=90)
results['SARIMA'] = evaluate_forecast(test['quantity'], pred_sarima, 'SARIMA')

# === 4. Prophet ===
prophet_df = train.reset_index().rename(columns={"timestamp": "ds", "quantity": "y"})
model_prophet = Prophet()
model_prophet.fit(prophet_df)
future = model_prophet.make_future_dataframe(periods=90)
forecast = model_prophet.predict(future)
pred_prophet = forecast.iloc[-90:]['yhat'].values
results['Prophet'] = evaluate_forecast(test['quantity'].values, pred_prophet, 'Prophet')

# === 5. XGBoost ===
df_features = df.dropna(subset=['lag_7', 'lag_14', 'lag_30', 'rolling_mean_7'])
xgb_features = ["lag_7", "lag_14", "lag_30", "rolling_mean_7"]

train_xgb = df_features[df_features['timestamp'] < test.index[0]]
test_xgb = df_features[df_features['timestamp'] >= test.index[0]]

model_xgb = XGBRegressor()
model_xgb.fit(train_xgb[xgb_features], train_xgb['quantity'])
pred_xgb = model_xgb.predict(test_xgb[xgb_features])
results['XGBoost'] = evaluate_forecast(test_xgb['quantity'], pred_xgb, 'XGBoost')

# === 6. Random Forest ===
model_rf = RandomForestRegressor()
model_rf.fit(train_xgb[xgb_features], train_xgb['quantity'])
pred_rf = model_rf.predict(test_xgb[xgb_features])
results['RandomForest'] = evaluate_forecast(test_xgb['quantity'], pred_rf, 'RandomForest')

# === 7. Visual Comparison ===
plt.figure(figsize=(12, 6))
plt.plot(test.index, test['quantity'], label='Actual', color='black')
plt.plot(test.index, pred_arima, label='ARIMA')
plt.plot(test.index, pred_sarima, label='SARIMA')
plt.plot(test.index, pred_prophet, label='Prophet')
plt.plot(test_xgb['timestamp'], pred_xgb, label='XGBoost')
plt.plot(test_xgb['timestamp'], pred_rf, label='RandomForest')
plt.title('Forecast Comparison (Last 90 Days)')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Results Table
results_df = pd.DataFrame(results, index=['RMSE', 'MAE']).T.sort_values('RMSE')
print("\nðŸ“Š Model Performance Comparison:")
print(results_df)
```
